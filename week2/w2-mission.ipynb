### 📌Q3. 강남지역에는 다른 지역에 비해 피부과나 성형외과가 많아 보입니다. 실제로 해당 지역에 피부과나 성형외과가 다른 지역에 비해 전체 병원 수 중에서 어느 정도의 비율을 차지하고 있는지 알아보겠습니다.

```python
df_seoul = df[df["시도명"]=="서울특별시"]
df_seoul_hospital = df_seoul[df_seoul["상권업종중분류명"] == "병원"]
df_seoul_target_hospital = df_seoul_hospital[df_seoul_hospital["상권업종소분류명"].str.contains("성형|피부")]

result = df_seoul_target_hospital.loc[:,"시군구명"].value_counts() / df_seoul_hospital.loc[:,"시군구명"].value_counts()
result.sort_values(ascending=False).round(2)
```

### 📌Q4. 평균 기대수명이 점점 길어지면서 실버 의료 산업도 주목받고 있습니다. 여러분은 실버 의료 산업과 관련된 스타트업에 취업했습니다. 지도를 시각화하여 '노인/치매병원'이 주로 어디에 위치하고 있는지를 찾아보고자 합니다.

```python
import folium

location = [df["위도"].mean(), df["경도"].mean()]
map = folium.Map(location=location, zoom_start=7)

df_old_man_hospital= df[df["상권업종소분류명"] == "노인/치매병원"]

for n in df_old_man_hospital.index :
    name = df_old_man_hospital.loc[n, "상호명"]
    adress = df_old_man_hospital.loc[n, "도로명주소"]
    popup = f"{name}-{adress}"
    
    hospital_location = [df_old_man_hospital.loc[n, "위도"], df_old_man_hospital.loc[n, "경도"]]
    folium.Marker(location=hospital_location, popup=popup).add_to(map)

map
    
```
